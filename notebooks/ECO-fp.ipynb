```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/zhutchens/Documents/Research/RESOLVE/polimera25_internal_upload/preupload_versions/eco_internal_112625.csv")
```

```python
df = df[(df.absrmag<=-17.33) & (df.g3grpcz > 3000) & (df.g3grpcz < 7000)]
```

```python
plt.figure(figsize=(11,4))
sc=plt.scatter(df.radeg * (24/360), df.dedeg, s=0.5, alpha=0.7, c=df.g3logmhvir, cmap='jet', rasterized=True)
cb=plt.colorbar(sc)
cb.set_label("Log Group Halo Mass", fontsize=10)
plt.xlim(130/15,238.5/15)
plt.ylim(-1,50)
plt.xlabel("RA [hr]", fontsize=12)
plt.ylabel("Decl. [deg]", fontsize=14)
plt.tight_layout()
plt.savefig("../figures/ECO-fp.png",dpi=300)
plt.show()
```
