```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pickle.load(open('../catalogs/LLvalues.pkl','rb'))
```

```python
bperp = data[0].astype(np.float64)
blos = data[1].astype(np.float64)
rss = data[2].astype(np.float64)
```

```python
optimal = np.unravel_index(np.argmin(rss, axis=None), rss.shape)
```

```python
bperp[optimal], blos[optimal]
```

```python
plt.figure(figsize=(5,5))
im=plt.imshow(rss, origin='lower')
plt.yticks(np.arange(0,bperp.shape[0]), labels=[f"{x:0.3f}" for x in bperp[:,0]])
plt.xticks(np.arange(0,blos.shape[0]), labels=[f"{x:0.3f}" for x in blos[0]], rotation=90)
plt.plot([optimal[1]], [optimal[0]], 'rx')
plt.show()
```
